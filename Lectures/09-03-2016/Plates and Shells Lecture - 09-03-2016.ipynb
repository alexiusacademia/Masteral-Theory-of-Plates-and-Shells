### Example Problem:
   Determine the deflection and stress in a very long and narrow rectangular plate if it is simply supported at edges <pre>y = 0 and y = b</pre>.
a) The plate is carrying a uniform loading expression 
\begin{equation*}
p(y) = P_o Sin\frac{\pi y}{b}
\end{equation*}
where the constant $P_o$ represents the intensity along the line passing thru $y = \frac{b}{2}$ parallel to x=axis.
$\nu = \frac{1}{3}$ , $t = 10mm$, $E = 200GPa$, $P_o = 10kPa$ and $b = 0.4m$.

<img src="figure1.jpg" />

### Solution:
Since the deflection is a function of 'y' only, the equation reduces to,
\begin{equation*}
\frac{P}{D} = \frac{\partial^4\omega}{\partial y^4}   or  
\frac{\partial^4\omega}{\partial y^4} = \frac{P}{D}  => Beam Equation
\end{equation*}

Substitute:<br>
\begin{equation*}P = P(y) = P_o sin \frac{\pi}{b} y \end{equation*}<br>
\begin{equation*}
\frac{d^4\omega}{dy^4} = \frac{P_o sin\frac{\pi}{b}y}{D}
\end{equation*}
<br>
\begin{equation*}
\frac{d^3\omega}{dy^3} = \frac{P}{D}\frac{b}{\pi}[-\cos \frac{\pi}{b}y] + C_1
\end{equation*}
<br>
\begin{equation*}
\frac{d^2\omega}{dy^2} = \frac{P}{D}\frac{b^2}{\pi^2}[-\sin \frac{\pi}{b}y] + C_1 y + C_2
\end{equation*}
<br>
\begin{equation*}
\frac{d\omega}{dy} = \frac{P}{D}\frac{b^3}{\pi^3}[\cos \frac{\pi}{b}y] + C_1*\frac{y^2}{2} + C_2 y+ C3
\end{equation*}
<br>
$$
\omega = \frac{P}{D}\frac{b^4}{\pi^4}[\sin \frac{\pi}{b}y] + C_1 \frac{y^3}{3} + C_2 \frac{y^2}{2} + C_3 y + C_4
$$

Apply boundary conditions:<br>
$ \omega = 0 $ at $y = 0$ and $y = b$

$$
\omega = \frac{P}{D} \frac{b^4}{\pi^4} sin \frac{\pi}{b}y
$$

#### Moment Values:
$$ M_x = -d(\nu \frac{\partial^2\omega}{\partial^2}) \quad since \frac{\partial^22\omega}{\partial x^2} = 0 \quad because \frac{\partial \omega}{\partial x} = 0\quad\quad \text{Slope along x-axis}$$ 
$$ M_x = -D \nu [\frac{P}{D}\frac{b^2}{\pi^2}(-\sin \frac{\pi}{b}y)] $$
$$ M_x = \nu P_o \frac{b^2}{\pi^2} \sin \frac{\pi}{b}y $$
$$ M_x = \frac{1}{3 \pi^2} P_o b^2 $$
<br>
$$ \sigma_x = \frac{12 M_x z}{t^3}\quad = \quad \frac{12 M_x (\frac{t}{2})}{t^3} $$


$$ M_y = -D \frac{\partial^2\omega}{\partial y^2} = -D [\frac{P b^2}{D \pi^2}(- sin \frac{\pi}{b}y)] $$
$$ M_y = \frac{P b^2}{\pi^2} sin \frac{\pi}{b}y \quad \text substitute \frac{b}{2} = y$$
$$ M_y = \frac{P b^2}{\pi^2} sin \frac{\pi}{2} $$

$$ \sigma_y = \frac{12 M_y z}{t^3} \quad z = \frac{t}{2} \text for max stress$$
